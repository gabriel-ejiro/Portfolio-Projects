```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#connect to website and pull in the data

url = 'https://en.wikipedia.org/wiki/List_of_countries_by_marriage_rate'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
# specify the data we are looking for
soup.find_all('table')[0]
soup.find('table', class_ ='wikitable sortable mw-datatable static-row-numbers')
table = soup.find('table')
print(table)
```

```python
# find table headers
world_titles = soup.find_all('th')
world_titles
```

```python
# get the table headers using a list comprehension
countries_title = [title.text.strip() for title in world_titles]
# this loops through and gets the title
 countries_title
```

```python
# create dataframe
df = pd.DataFrame(columns = countries_title)
df
```

```python
# create loop to extract the table data and insert into dataframe
column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data

df
```

```python
# export to csv
df.to_csv("countries_marriage_rate.csv", index=False)
```
